```python
import pandas as pd
import matplotlib.pyplot as plt

CT18 = pd.read_csv('Data/Carbon_Tax_18.csv')
CT18 = CT18.add_suffix('_CT18')
CT20 = pd.read_csv('Data/Carbon_Tax_20.csv')
CT20 = CT20.add_suffix('_CT20')
CT22 = pd.read_csv('Data/Carbon_Tax_22.csv')
CT22 = CT22.add_suffix('_CT22')
CT24 = pd.read_csv('Data/Carbon_Tax_24.csv')
CT24 = CT24.add_suffix('_CT24')
CT26 = pd.read_csv('Data/Carbon_Tax_26.csv')
CT26 = CT26.add_suffix('_CT26')

UP5 = pd.read_csv('Data/Uniform_Price.csv')
UP5 = UP5.add_suffix('_UP5')
DP5 = pd.read_csv('Data/Discriminatory_Price.csv')
DP5 = DP5.add_suffix('_DP5')
SP5 = pd.read_csv('Data/Second_Price.csv')
SP5 = SP5.add_suffix('_SP5')

AVG = pd.read_csv('Data/Uniform_Price_Average.csv')
AVG = AVG.add_suffix('_AVG')
MCA = pd.read_csv('Data/Uniform_Price_McAfee.csv')
MCA = MCA.add_suffix('_MCA')
VCG = pd.read_csv('Data/Uniform_Price_VCG.csv')
VCG = VCG.add_suffix('_VCG')



dfList = [CT18,CT20,CT22,CT24,CT26,UP5,DP5,SP5,AVG,MCA,VCG]

df = pd.concat(dfList)

df.columns
```

```python
CT18latex = CT18[["Emissions_CT18", 'Top Fifty Polluters Supply_CT18', "Average Polluters Supply_CT18", "Top Fifty Efficient Supply_CT18"]]
CT18latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
CT18latex = CT18latex.iloc[::100]
CT18latex.to_csv("Tables/CT18Latex.csv")

CT20latex = CT20[["Emissions_CT20", 'Top Fifty Polluters Supply_CT20', "Average Polluters Supply_CT20", "Top Fifty Efficient Supply_CT20"]]
CT20latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
CT20latex = CT20latex.iloc[::100]
CT20latex.to_csv("Tables/CT20Latex.csv")

CT22latex = CT22[["Emissions_CT22", 'Top Fifty Polluters Supply_CT22', "Average Polluters Supply_CT22", "Top Fifty Efficient Supply_CT22"]]
CT22latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
CT22latex = CT22latex.iloc[::100]
CT22latex.to_csv("Tables/CT22Latex.csv")

CT24latex = CT24[["Emissions_CT24", 'Top Fifty Polluters Supply_CT24', "Average Polluters Supply_CT24", "Top Fifty Efficient Supply_CT24"]]
CT24latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
CT24latex = CT24latex.iloc[::100]
CT24latex.to_csv("Tables/CT24Latex.csv")

CT26latex = CT26[["Emissions_CT26", 'Top Fifty Polluters Supply_CT26', "Average Polluters Supply_CT26", "Top Fifty Efficient Supply_CT26"]]
CT26latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
CT26latex = CT26latex.iloc[::100]
CT26latex.to_csv("Tables/CT26Latex.csv")

UP5latex = UP5[["Emissions_UP5", 'Top Fifty Polluters Supply_UP5', "Average Polluters Supply_UP5", "Top Fifty Efficient Supply_UP5"]]
UP5latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
UP5latex = UP5latex.iloc[::100]
UP5latex.to_csv("Tables/UP5Latex.csv")

DP5latex = DP5[["Emissions_DP5", 'Top Fifty Polluters Supply_DP5', "Average Polluters Supply_DP5", "Top Fifty Efficient Supply_DP5"]]
DP5latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
DP5latex = DP5latex.iloc[::100]
DP5latex.to_csv("Tables/DP5Latex.csv")

SP5latex = SP5[["Emissions_SP5", 'Top Fifty Polluters Supply_SP5', "Average Polluters Supply_SP5", "Top Fifty Efficient Supply_SP5"]]
SP5latex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
SP5latex = SP5latex.iloc[::100]
SP5latex.to_csv("Tables/SP5Latex.csv")

AVGlatex = AVG[["Emissions_AVG", 'Top Fifty Polluters Supply_AVG', "Average Polluters Supply_AVG", "Top Fifty Efficient Supply_AVG"]]
AVGlatex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
AVGlatex = AVGlatex.iloc[::100]
AVGlatex.to_csv("Tables/AVGLatex.csv")

MCAlatex = MCA[["Emissions_MCA", 'Top Fifty Polluters Supply_MCA', "Average Polluters Supply_MCA", "Top Fifty Efficient Supply_MCA"]]
MCAlatex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
MCAlatex = MCAlatex.iloc[::100]
MCAlatex.to_csv("Tables/MCALatex.csv")

VCGlatex = VCG[["Emissions_VCG", 'Top Fifty Polluters Supply_VCG', "Average Polluters Supply_VCG", "Top Fifty Efficient Supply_VCG"]]
VCGlatex.columns = ["Emissions", "Top-50 Polluters Supply", "Avg. Polluters Supply", "Top-50 Efficient Supply"]
VCGlatex = VCGlatex.iloc[::100]
VCGlatex.to_csv("Tables/VCGLatex.csv")
```

```python
df['Emissions Rolling_CT18'] = df['Emissions_CT18'].rolling(window=100).mean()
df['Emissions Rolling_CT20'] = df['Emissions_CT20'].rolling(window=100).mean()
df['Emissions Rolling_CT22'] = df['Emissions_CT22'].rolling(window=100).mean()
df['Emissions Rolling_CT24'] = df['Emissions_CT24'].rolling(window=100).mean()
df['Emissions Rolling_CT26'] = df['Emissions_CT26'].rolling(window=100).mean()


df.plot(y=['Emissions Rolling_CT18','Emissions Rolling_CT20','Emissions Rolling_CT22','Emissions Rolling_CT24','Emissions Rolling_CT26'], color=["#a9ffc0","#7fc992", "#559364",  "#31653d", "#002607"])
plt.title('Total Emissions (Rolling Average)')
plt.xlabel('Ticks')
plt.ylabel('Emissions (tCO2)')
plt.legend(['18 Tax', '20 Tax', '22 Tax', '24 Tax', '26 Tax'])
fig1 = plt.gcf()
plt.show()

fig1.savefig("Graphs/CarbonTax_TotalEmissions.png")
```
